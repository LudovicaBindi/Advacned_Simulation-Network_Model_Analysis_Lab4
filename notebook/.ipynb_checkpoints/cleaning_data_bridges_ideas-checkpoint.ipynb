```python
import pandas as pd
import numpy as np
from IPython.display import Image
```

```python
bridges = pd.read_excel("../data/BMMS_overview.xlsx")
```

```python
bridges.head()
```

```python
n1_db = bridges.loc[bridges['road'] == 'N1'].sort_values(by=['lat'])
```

```python
n2_db = bridges.loc[bridges['road'] == 'N2'].sort_values(by=['lat'])
```

```python
def db_to_array(db):
    result = []
    for index, row in db.iterrows():
        result.append([row['LRPName'], row['lat'], row['lon']])
    return result
```

```python
n1_ar = db_to_array(n1_db)
n2_ar = db_to_array(n2_db)
```

Idea: find the bridges that are close to one another because it means that these bridges belonging to different roads are used to indeed connect the roads.</br>

Basic algorithm: check all the possible combinations of LRPs between the two lists ( O(n^2) )</br>
Ideas to improve the velocity:
* break after we find a match
* sort the values (this could be done in the dataframe easily, before we create the arrays)

```python
# basic algorithm using '='
for n1 in n1_ar:
    for n2 in n2_ar:
        if n1[1] == n2[1] and n1[2] == n2[2]:
            print('N1: ' + n1[0] + ', N2: ' + n2[0])
            break # since we found a match there is no need to keep looking for others for this N1 LRP
```

```python
def check_similarity(n1, n2):
    if n1[1] > n2[1]:
        dif_lat_per = ((n1[1] - n2[1]) / n1[1])
    else:
        dif_lat_per = ((n2[1] - n1[1]) / n2[1])
    if n1[2] > n2[2]:
        dif_lon_per = ((n1[2] - n2[2]) / n1[2])
    else:
        dif_lon_per = ((n2[2] - n1[2]) / n2[2])
    if dif_lat_per < 0.001 and dif_lon_per < 0.001:
        return True
    else:
        return False

# with the use of (global) constants we don't need to hardcode the percentages (0.01)

# this algorithm could be further improved by finding the right constant by looking at how the distance in lon and lan is in Km
```

```python
# basic algorithm using similarity check
for n1 in n1_ar:
    for n2 in n2_ar:
        if check_similarity(n1, n2) == True:
            print('N1: ' + n1[0] + ', N2: ' + n2[0])
            break # since we found a match there is no need to keep looking for others for this N1 LRP
```

Let's implement the full version using the basic algorithm:

```python
# get all the N roads we are interest in
roads = bridges.loc[bridges['road'].str.startswith('N')]['road'].unique()
roads = np.delete(roads, [0, 1])
```

```python
# check intersection with N1
for road in roads:
    # get array of LRPs in this road
    road_db = bridges.loc[bridges['road'] == road].sort_values(by=['lat'])
    road_ar = db_to_array(road_db)
    # basic algorithm using similarity check
    for bridge in road_ar:
        for n1 in n1_ar:
            if check_similarity(bridge, n1) == True:
                print(road + ': ' + bridge[0] + ', N1: ' + n1[0])
                break # since we found a match there is no need to keep looking for others for this N1 LRP
```

```python
# check intersection with N2
for road in roads:
    # get array of LRPs in this road
    road_db = bridges.loc[bridges['road'] == road].sort_values(by=['lat'])
    road_ar = db_to_array(road_db)
    # basic algorithm using similarity check
    for bridge in road_ar:
        for n2 in n2_ar:
            if check_similarity(bridge, n2) == True:
                print(road + ': ' + bridge[0] + ', N2: ' + n2[0])
                break # since we found a match there is no need to keep looking for others for this N1 LRP
```

It takes a couple of seconds to run each part, so it's not too bad for this limited case. If we were to consider more roads, then this would be a huge limitation.
